```python
def findFrequencyFiles(urls):
    counter_dict = {}
    for url in urls:
        str_start = url.rfind("/")
        filename = url[str_start+1:]
        if filename in counter_dict:
#             print(filename)
            counter_dict[filename] += 1
        else:
            counter_dict[filename] = 1
        print(url)
        
#     print(counter_dict)
    print(sorted(counter_dict.items(), key=lambda d: d[1], reverse=True)[:3])
```

```python
urls = [
"http://www.google.com/a.txt",
"http://www.google.com.tw/a.txt",
"http://www.google.com/download/c.jpg",
"http://www.google.co.jp/a.txt",
"http://www.google.com/b.txt",
"https://facebook.com/movie/b.txt",
"http://yahoo.com/123/000/c.jpg",
"http://gliacloud.com/haha.png",
]
findFrequencyFiles(urls)
```

```python
from bs4 import BeautifulSoup 
import requests 
import time
```

```python
def get_web_page(url):
    resp = requests.get(
        url=url,
#         cookies={'over18': '1'}
    )
    if resp.status_code != 200:
        print('Invalid url:', resp.url)
        return None
    else:
        return resp.text
    
def get_articles(dom):
    soup = BeautifulSoup(dom, 'html.parser')

    index = 0
    articles = []  # 儲存取得的文章資料
    divs = soup.find_all('div', 'r-ent')
    for d in divs:
#         print(d.find('div', 'date').string)
#         if d.find('div', 'date').string == date:  # 發文日期正確
        # 取得推文數
        push_count = 0
        if d.find('div', 'nrec').string:
            try:
                push_count = int(d.find('div', 'nrec').string)  # 轉換字串為數字
            except ValueError:  # 若轉換失敗，不做任何事，push_count 保持為 0
                pass
#         div class="author"
#         if d['class'] == "auther":
# <div class="author">ReiKuromiya</div>
#         print(d.find('class') == "author")
        
        # 取得文章連結及標題
        if d.find('a'):  # 有超連結，表示文章存在，未被刪除
            href = d.find('a')['href']
            title = d.find('a').string
            articles.append({
                'date':d.find('div', 'date').string,
                'title': title,
                
#                     'href': href,
#                     'push_count': push_count
            })
            
    divs = soup.find_all("div", class_="author")
    for d in divs:
        articles[index]["author"] = d.string
        index+=1
#         print(d.string)
    return articles
```

```python
page = get_web_page('https://www.ptt.cc/bbs/Beauty/index.html')
if page:
    current_articles = get_articles(page)
    for post in current_articles:
        print(post)
```
